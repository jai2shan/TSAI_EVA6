```python
from cifar10.dataloader import *
from cifar10.models import *
from cifar10.utils import *
from cifar10.GradCam import *
```

```python
data = cifar_dataloader(1000)
trainloader, testloader = data()
```

```python
trainNtest = TrainTest(ResNet18(), trainloader, testloader, lr = 0.05)
trainNtest(20)
```

```python
im_pred = Misclassified_Images(trainNtest.model, data, testloader,device = "cuda")
```

```python
plot_Misclassified(im_pred, data)
```

```python
GradCamDisplay(trainNtest.model, im_pred['Correct'], data.classes, trainNtest.device, True)
```

```python
GradCamDisplay(trainNtest.model, im_pred['Wrong'], data.classes, trainNtest.device, False)
```
